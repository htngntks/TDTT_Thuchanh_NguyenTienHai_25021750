```python
#W4A8
def palindrome(num):
  rev = 0
  while(num > 0):
    rev = rev * 10 + num % 10
    num //= 10
  return rev

while True:
  try:
    n = int(input("Nhập một số nguyên dương: "))
    if n > 0:
      break
    else:
      print("n phải là số nguyên dương.")
  except ValueError:
    print("Vui lòng nhập một số nguyên.")
step = 0
while palindrome(n) != n:
  n += palindrome(n)
  step += 1

print(step, n)
```

```python
#W4A9
def distinct(num):
  num = abs(num)
  if num == 0:
      return True
  digits = []
  while num > 0:
    digit = num % 10
    if digit in digits:
      return False
    digits.append(digit)
    num //= 10
  return True

while True:
  try:
    n = int(input("Nhập một số nguyên dương: "))
    if n > 0:
      break
    else:
      print("n phải là số nguyên dương.")
  except ValueError:
    print("Vui lòng nhập một số nguyên.")

for i in range(1, int(n ** 0.5) + 1):
  if distinct(i):
    print(i * i, end=' ')
```

```python
#W4A30
def upper_letter(s):
  sum = 0
  for letter in s:
    if letter.isupper():
      sum += 1
  return sum
def lower_letter(s):
  sum = 0
  for letter in s:
    if letter.islower():
      sum += 1
  return sum
def digit(s):
  sum = 0
  for letter in s:
    if letter.isdigit():
      sum += 1
  return sum

s = input()
print(upper_letter(s), lower_letter(s), digit(s))
```

```python
#W4A32
def upper_char(st):
  for letter in st:
    if letter.isupper():
      return True
  return False
def lower_char(st):
  for letter in st:
    if letter.islower():
      return True
  return False
def digit_char(st):
  for letter in st:
    if letter.isdigit():
      return True
  return False
def special_char(st):
    for letter in st:
        if not letter.isalnum():
            return True
    return False
chuoi = input()
print(special_char(chuoi) and upper_char(chuoi) and lower_char(chuoi) and digit_char(chuoi) and len(chuoi) >= 6)
```
