```python
#W11A1
def reverse(path:str) -> int:
  try:
    with open(path, 'r') as f:
      return int(f.read().strip()[::-1])
  except:
    print("Err")
```

```python
#W11A4
def areAnagrams(path1: str, path2: str):
  def sorted_chr(text):
    return sorted([chr.lower() for chr in text if chr.isalpha()])
  with open(path1, 'r') as f1, open(path2, 'r') as f2:
    s1 = f1.read()
    s2 = f2.read()
    return sorted_chr(s1) == sorted_chr(s2)
```

```python
#W11A12
def maximumProduct(path: str) -> int:
  with open(path, 'r') as f:
    s = f.read().split()
  nums = [int(x) for x in s[1:]]
  nums.sort()
  return max(nums[0] * nums[1] * nums[-1], nums[-1] * nums[-2] * nums[-3])

```

```python
#W11A13
def moveZeroes (fileName: str) -> list:
  with open(fileName, 'r') as f:
    nums = [int(x) for x in f.read().split()]
  for i in range(len(nums) - 1, -1, -1):
    if nums[i] == 0:
      nums.pop(i)
      nums.append(0)
  return nums
```

```python
#W11A14
import csv
def averageTime(path: str):
  with open(path, 'r') as f:
    data = csv.reader(f)
    next(data)
    sum = 0
    count = 0
    for row in data:
      try:
        sum += float(row[1])
        count += 1
      except:
        continue
  if count == 0:
    return "0.00"
  return '{:.2f}'.format(round(sum/count, 2))
```

```python
#11A16
import math
path = input()
with open(path, 'r') as f:
  lines = [l.strip() for l in f if l.strip()]
for i in range (0, len(lines), 2):
  shape_type = lines[i]
  data = lines[i+1]

  if "SQUARE" in shape_type:
    side = float(data.split(':')[1])
    res = side * 4
  elif "RECTANGLE" in shape_type:
    width, height = float(data.split()[1]), float(data.split()[3])
    res = (width + height) * 2
  elif "CIRCLE" in shape_type:
    radius = float(data.split(':')[1])
    res = 2 * math.pi * radius
  print('{:.2f}'.format(round(res, 2)))
```

```python
#W11A17
def productExceptSelf (path: str) -> list[int]:
  res = []
  with open(path, 'r') as f:
    nums = [int(x) for x in f.read().split()]
    tich = 1
    for x in nums:
      tich *= x
    for x in nums:
      res.append(tich // x)
  return res
```

```python
#W11A18
def estimatedTime(path: str):
  with open(path, 'r') as f:
    data = f.read().splitlines()
    x1, y1 = float(data[0].split()[2]), float(data[0].split()[4])
    x2, y2 = float(data[1].split()[2]), float(data[1].split()[4])
    vt = float(data[2].split()[1])
  s = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
  t = s / vt
  return '{:.2f}'.format(round(t, 2))
```

```python
#W11A20
def imageSmoother(path: str):
  with open(path, 'r') as f:
    lines = f.read().splitlines()
  M = int(lines[0])
  img = []
  for i in range (1, M+1):
    row = list(map(int, lines[i].split()))
    img.append(row)
  res = [[0] * M for x in range(M)]

  for r in range(M):
    for c in range(M):
      sum = 0
      count = 0
      for i in range(r-1,r+2):
        for j in range(c-1, c+2):
          if 0 <= i < M and 0 <= j < M:
            sum += img[i][j]
            count += 1
      res[r][c] = sum // count
  for row in res:
    print(*row)
```
