```python
#W5A1
def Max_Of_Two(a, b):
    if (a >= b):
        return a
    else:
        return b
a, b = map(int, input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a, b):
    a, b = b, a
    return a, b

a, b = map(int, input().split())
print(swap(a, b))
```

```python
#W5A3
def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0:
        return False
    else:
        for i in range (3, int(n**0.5)+1, 2):
            if (n % i == 0): return False
    return True
n = int(input())
print(is_prime(n))
```

```python
#W5A4
def perf_num(n):
    sum = 0
    i = 1
    while (i <= n/2):
        if(n%i==0): 
            sum+=i
        i+=1
    return (sum == n)
n = int(input())
print(perf_num(n))
```

```python
#W5A5
def find_num(numlist, k):
    i = 0
    while(i < len(numlist) and numlist[i] != k):
        i+=1
    if i == len(numlist):
        return -1
    return i + 1
lst = list(map(int, input().split()))
k = int(input())
print(find_num(lst, k))

```

```python
#W5A6
def giaithua(n):
    tich = 1
    i = 1
    while(i <= n):
        tich*=i
        i+=1
    return tich
n = int(input())
print(giaithua(n))
```

```python
#W5A7
def gialap(a, b , chr):
    match(chr):
        case '+': return format(a + b, ",.2f")
        case '-': return format(a - b, ",.2f")
        case '*': return format(a * b, ",.2f")
        case '/': 
            if b == 0: return "Khong the thuc hien phep chia cho 0"
            return format(a / b, ",.2f")
num1, operator, num2 = map(str, input().split())
print(gialap(int(num1), int(num2), operator))
```

```python
#W5A8
def hamming(a, b):
    khoang_cach= 0
    while a > 0 or b > 0:
        lastbitA = a % 2
        lastbitB = b % 2
        if lastbitA != lastbitB:
            khoang_cach+=1
        a //= 2
        b //= 2
    return khoang_cach

a, b = int(input())
print(hamming(a, b))
```

```python
#W5A9
def sumDigit(n):
    sum = 0
    while(n > 0):
        sum += n%10
        n//=10
    return sum
n = int(input())
print(sumDigit(n))
```

```python
#W4A10
def dang_cau(a, b):
    lst=[]
    for i in range (len(a)):
        lst.append((a[i], b[i]))
    for i in range (len(lst)):
        for j in range (len(lst)):
            if (j!=i and ((lst[i][0] == lst[j][0] and lst[i][1] != lst[j][1]) or (lst[i][0] != lst[j][0] and lst[i][1] == lst[j][1]))):
                return False
    return True
a, b = map(str, input().split())
print(dang_cau(a, b))
```
