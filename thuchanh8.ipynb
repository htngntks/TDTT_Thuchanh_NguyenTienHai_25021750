```python
#W8A1
class Cylinder:
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height
  def area(self):
    return round(2 * 3.14 * self.radius * self.height + 2 * 3.14 * self.radius ** 2, 2)
  def volume(self):
    return round(3.14 * self.radius ** 2 * self.height, 2)
a, b = map(float, input().split())
c = Cylinder(a, b)
print(c.area())
print(c.volume())
```

```python
#W8A2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year
  def check(self):
    res = True
    months = [[1,3,5,7,8,10,12],[2],[4,6,9,11]]
    if self.month not in range(1,13):
      res = False
    if self.month in months[0]:
      if self.day not in range(1,32):
        res = False
    elif self.month in months[1]:
      if self.year % 400 == 0 or (self.year % 100 != 0 and self.year % 4 == 0):
        if self.day not in range(1,30):
          res = False
      else:
        if self.day not in range(1,29):
          res = False
    elif self.month in months[2]:
      if self.day not in range(1,31):
        res = False
    if res == False:
      print('INVALID')
    return res
  def nextday(self):
    months = [[1,3,5,7,8,10,12],[2],[4,6,9,11]]
    if self.month in months[0]:
      if self.day == 31:
        self.day = 1
        if self.month == 12:
          self.month = 1
          self.year += 1
        else:
          self.month += 1
      else:
        self.day += 1
    if self.month in months[1]:
      if self.year % 400 == 0 or (self.year % 100 != 0 and self.year % 4 == 0):
        if self.day == 29:
          self.day = 1
          self.month += 1
        else:
          self.day += 1
      else:
        if self.day == 28:
          self.day = 1
          self.month += 1
        else:
          self.day += 1
    if self.month in months[2]:
      if self.day == 30:
        self.day = 1
        self.month += 1
      else:
        self.day += 1
    return f'{self.day}/{self.month}/{self.year}'
day, month, year = map(int, input().split('/'))
d = Date(day, month, year)
if d.check():
  print(d.nextday())
```

```python
#W8A3
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def is_O(self):
    return (self.x == 0 and self.y == 0)
  def on_Ox(self):
    return (self.x != 0 and self.y == 0)
  def on_Oy(self):
    return (self.x == 0 and self.y != 0)
  def distance(self, other):
    return '{:.2f}'.format(((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5)
a, b = map(int, input().split())
O = Point(0,0)
M = Point(a, b)
print(M.is_O())
print(M.on_Ox())
print(M.on_Oy())
print(M.distance(O))
```

```python
#W8A4
class Caculator:
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2
  def add(self):
    return self.num1 + self.num2
  def sub(self):
    return self.num1 - self.num2
  def product(self):
    return self.num1 * self.num2
  def divide(self):
    if self.num2 == 0:
      return 'Err'
    return self.num1 / self.num2
  def power(self):
    return self.num1 ** self.num2
  def mod(self):
    if self.num2 == 0:
      return 'Err'
    return self.num1 % self.num2
  def set_numbers(self, num1, num2):
    self.num1 = num1
    self.num2 = num2
    print(f"Số mới: {self.num1}, {self.num2}")
def main():
  a, b = map(float, input().split())
  calc = Caculator(a, b)
  while True:
    print("\n-----------MENU-----------")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia lấy dư")
    print("7. Nhập bộ số mới")
    print("--------------------------")

    choice = input("Nhập lựa chọn của bạn: ")
    if choice == '1':
      print(f"Kết quả: {calc.add()}")
    elif choice == '2':
      print(f"Kết quả: {calc.sub()}")
    elif choice == '3':
      print(f"Kết quả: {calc.product()}")
    elif choice == '4':
      print(f"Kết quả: {calc.divide()}")
    elif choice == '5':
      print(f"Kết quả: {calc.power()}")
    elif choice == '6':
      print(f"Kết quả: {calc.mod()}")
    elif choice == '7':
      a, b = map(float, input().split())
      calc.set_numbers(a, b)
    else:
      print("Err")
    exit = input("\nBạn có muốn thoát?: ")
    if exit in ['yes','YES', 'Yes']:
      break
main()
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        item = {"name": name, "price": price}
        self.items.append(item)

    def delete_item(self, name):
        found = False
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                found = True
                break
        if not found:
            print("Không tìm thấy sản phẩm")

    def check_empty(self):
        return len(self.items) == 0

    def total(self):
        total_price = 0
        for item in self.items:
            total_price += item["price"]
        return total_price

    def show_items(self):
        if self.check_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.items:
                print(f"Tên: {item['name']}, Giá: {item['price']}")

    def delete_all(self):
        self.items.clear()

cart = ShoppingCart()

while True:
    print("\n--- MENU ---")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Kiểm tra giỏ rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ")

    choice = input("Nhập lựa chọn: ")

    if choice == '2':
        name = input("Nhập tên: ")
        price = float(input("Nhập giá: "))
        cart.add_item(name, price)
    elif choice == '3':
        name = input("Nhập tên cần xóa: ")
        cart.delete_item(name)
    elif choice == '4':
        if cart.check_empty():
            print("Giỏ hàng đang rỗng")
        else:
            print("Giỏ hàng không rỗng")
    elif choice == '5':
        print(f"Tổng tiền: {cart.total()}")
    elif choice == '6':
        cart.show_items()
    elif choice == '7':
        cart.delete_all()
        print("Đã xóa toàn bộ giỏ hàng")

    ask = input("Có muốn thoát khỏi chương trình không? ")
    if ask == "yes":
        break
```

```python
#W8A6
class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width
  def area(self):
    return self.length * self.width
  def perimeter(self):
    return 2 * (self.length + self.width)
  def scale(self, k):
    self.length *= k
    self.width *= k
w, h, k = map(int, input().split())
rec1 = Rectangle(w, h)
rec1.scale(k)
print("{:.2f} {:.2f}".format(rec1.area(), rec1.perimeter()))
```

```python
#W8A7
def gcd(a, b):
  a, b = abs(a), abs(b)
  if b == 0:
    return a
  return gcd(b, a%b)
class Fraction:
    def __init__(self, num, den):
      if den < 0:
          num = -num
          den = -den
      self.num = num//gcd(num, den)
      self.den = den//gcd(num, den)

    def __str__(self):
      return f"{self.num}/{self.den}"

    def add(self, other):
      return Fraction(self.num * other.den + self.den * other.num, self.den * other.den)
    def sub(self, other):
      return Fraction(self.num * other.den - self.den * other.num, self.den * other.den)
    def mul(self, other):
      return Fraction(self.num * other.num, self.den * other.den)
    def div(self, other):
      return Fraction(self.num * other.den, self.den * other.num)

a, b, op, c, d = map(str, input().split())
a, b, c, d = int(a), int(b), int(c), int(d)
frac1 = Fraction(a, b)
frac2 = Fraction(c, d)
if op == '+':
    print(frac1.add(frac2))
elif op == '-':
    print(frac1.sub(frac2))
elif op == '*':
    print(frac1.mul(frac2))
elif op == '/':
    print(frac1.div(frac2))
```

```python
#W8A8
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance
  def deposit(self, amount):
    self.balance += amount
  def withdraw(self, amount):
      self.balance -= amount
  def __str__(self):
    return f"{self.owner}: {self.balance}"
name, balance = input().split()
n = int(input())
balance = float(balance)
account = BankAccount(name, balance)
for i in range(n):
  a, b = input().split()
  b = float(b)
  if a == "DEPOSIT":
    account.deposit(b)
  else:
    account.withdraw(b)
print('{:.2f}'.format(account.balance))
```

```python
#W8A9
class Student:
  def __init__(self, name):
    self.name = name
    self.scores = []
  def add_score(self, subject, score):
    self.scores.append(score)
  def avg(self):
    if not self.scores:
      return 0.00
    return sum(self.scores) / len(self.scores)
  def rank(self):
    avg_sc = self.avg()
    if avg_sc >= 8:
      return "Excellent"
    elif avg_sc >= 6.5:
      return "Good"
    elif avg_sc >= 5:
      return "Average"
    else:
      return "Poor"
name = input().strip()
n = int(input())
student = Student(name)
for i in range (n):
  sub, sc = map(str, input().split())
  sc = float(sc)
  student.add_score(sub, sc)
print(f"{student.name} {student.avg():.2f} {student.rank()}")
```

```python
#W8A10
class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = int(year)

class Library:
  def __init__(self):
    self.books = []
  def add(self, book):
    self.books.append(book)
  def count_by_author(self, name):
    return sum(1 for b in self.books if b.author == name)
  def find_by_year(self, y):
    return sum(1 for b in self.books if b.year == y)
n = int(input())
lib = Library()
for i in range(n):
  lenh, b = input().split(' ', 1)
  if lenh == "ADD":
    t, a, y = b.split(";")
    lib.add(Book(t, a, y))
  elif lenh == "COUNT":
    print(lib.count_by_author(b))
  elif lenh == "COUNTYEAR":
    print(lib.find_by_year(int(b)))
```

```python
#W8A11
import math

class TinhGoc:
  def __init__(self):
    self.data = []
    for i in range(4):
      line = input().split()
      self.data.extend(map(float, line))

  def tinh_goc(self, p):
    ab = (p[3]-p[0], p[4]-p[1], p[5]-p[2])
    ac = (p[6]-p[0], p[7]-p[1], p[8]-p[2])
    bc = (p[6]-p[3], p[7]-p[4], p[8]-p[5])
    bd = (p[9]-p[3], p[10]-p[4], p[11]-p[5])

    n1 = (ab[1]*ac[2] - ab[2]*ac[1], ab[2]*ac[0] - ab[0]*ac[2], ab[0]*ac[1] - ab[1]*ac[0])
    n2 = (bc[1]*bd[2] - bc[2]*bd[1], bc[2]*bd[0] - bc[0]*bd[2], bc[0]*bd[1] - bc[1]*bd[0])

    dot = abs(n1[0]*n2[0] + n1[1]*n2[1] + n1[2]*n2[2])
    l1 = math.sqrt(n1[0]**2 + n1[1]**2 + n1[2]**2)
    l2 = math.sqrt(n2[0]**2 + n2[1]**2 + n2[2]**2)

    return math.degrees(math.acos(min(dot / (l1 * l2), 1.0)))

  def chay(self):
    for i in range(0, len(self.data), 12):
      print(f"{self.tinh_goc(self.data[i:i+12]):.2f}")

app = TinhGoc()
app.chay()
```

```python
#W8A12
class Complex:
  def __init__(self, re, im):
    self.re = re
    self.im = im
  def __str__(self):
    return f"{self.re} {self.im}"
  def mul(self, other):
    return Complex(round(self.re * other.re - self.im * other.im), round(self.re * other.im + self.im * other.re))
a, b = map(float, input().split())
c, d = map(float, input().split())
c1 = Complex(a, b)
c2 = Complex(c, d)
print(c1.mul(c2))
```

```python
#W8A13
class Triangle:
  def __init__(self, edge1, edge2, edge3):
    self.edge1 = edge1
    self.edge2 = edge2
    self.edge3 = edge3
  def check(self):
    return self.edge1 > 0 and self.edge2 > 0 and self.edge3 > 0 and self.edge1 + self.edge2 > self.edge3 and self.edge1 + self.edge3 > self.edge2 and self.edge2 + self.edge3 > self.edge1
  def getArea(self):
    p = (self.edge1 + self.edge2 + self.edge3) / 2
    return "{:.2f}".format((p * (p - self.edge1) * (p - self.edge2) * (p - self.edge3))**0.5, 2) if self.check() else "invalid"
a, b, c = map(int, input().split())
tamgiac = Triangle(a, b, c)
print(tamgiac.getArea())
```
