```python
#W6A2
def tong(lst, n):
  tong = lst[0]
  for i in range (1, n):
    tong += lst[i]
  return tong
lst = list(map(int, input().split()))
res = []
for i in range(len(lst)):
    res.append(tong(lst, i+1))
print(res)
```

```python
#W6A4
def tao_dict(lst):
  dct = {}
  for cap in lst:
    key, value = cap.split(":", 1)
    if key in dct:
      dct[key].append(value)
    else:
      dct[key] = [value]
  return dct
lst = list(map(str, input().split()))
print(tao_dict(lst))
```

```python
#W6A5
def dem_so(lst):
  duong = 0
  am = 0
  khong = 0
  for i in lst:
    if i > 0:
      duong += 1
    elif i < 0:
      am += 1
    else:
      khong += 1
  res = {'positives': duong, 'negatives': am, 'zeros': khong}
  return res
nums = list(map(int, input().split()))
print(dem_so(nums))
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())

for k in d2:
    if k in d1:
        d1[k] = d1[k] + d2[k]
    else:
        d1[k] = d2[k]
keys = list(d1.keys())
n = len(keys)
for i in range(n):
    for j in range(0, n - i - 1):
        if keys[j] > keys[j + 1]:
            temp = keys[j]
            keys[j] = keys[j + 1]
            keys[j + 1] = temp
ket_qua = {}
for k in keys:
    ket_qua[k] = d1[k]
print(ket_qua)
```

```python
#W6A10
lst = list(map(int, input().split()))
k = int(input())
ket_qua_set = set()
n = len(lst)
for i in range(n):
  for j in range(i + 1, n):
    if lst[i] + lst[j] == k:
      pair = (lst[i], lst[j])
      ket_qua_set.add(pair)
print(sorted(list(ket_qua_set)))
```

```python
#W6A19
def maxvalue(lst):
  if len(lst) < 1:
    return -1
  max_value = lst[0]
  index = 0
  for i in range(1, len(lst)):
    if lst[i] > max_value:
      max_value = lst[i]
      index = i
  return index + 1
nums = list(map(int, input().split()))
print(maxvalue(nums))
```

```python
#W6A20
def find_nums(lst, k):
  for i in range(len(lst)):
    if lst[i] == k:
      return i + 1
  return -1
lst = list(map(int, input().split()))
k = int(input())
print(find_nums(lst, k))
```
