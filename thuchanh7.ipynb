```python
#W7A1
def binary_search(arr, left, right, target):
  mid = (left + right) // 2
  if left > right:
    return -1
  elif arr[mid] == target:
    return mid
  elif arr[mid] < target:
    return binary_search(arr, mid + 1, right, target)
  else:
    return binary_search(arr, left, mid - 1, target)
lst = list(map(int, input().split()))
target = int(input())
print(binary_search(lst, 0, len(lst) - 1, target))
```

```python
#W7A2
def count_occurrences(arr, x) -> int:
  count = 0
  for i in arr:
    if i == x:
      count += 1
  return count
lst = list(map(int, input().split()))
x = int(input())
print(count_occurrences(lst, x))
```

```python
#W7A3
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    swapped = False
    for j in range(0, n - i - 1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swapped = True
    if not swapped:
      break
def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_index = i
    for j in range(i+1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
arr = list(map(int, input().split()))
selection_sort(arr)
print(*arr)
```

```python
#W7A7
def lis(nums):
  n = len(nums)
  res = [1] * n
  for i in range(n):
      for j in range(i):
          if nums[i] > nums[j] and res[i] < res[j] + 1:
              res[i] = res[j] + 1
  return max(res)
nums = eval(input())
print(lis(nums))
```

```python
#W7A8
interval = eval(input())
queries = eval(input())
def solve(nums, queries):
  res = []
  for i in nums:
    if i[0] <= queries <= i[1]:
      res.append(i[1]-i[0]+1)
  if len(res) == 0: return -1
  else: return min(res)
kq = []
for i in queries:
  kq.append(solve(interval, i))
print(kq)
```

```python
#W7A10
nums = eval(input())
m = int(input())
max_sum = 0
curr = 0
prefixes = [0]
for i in nums:
  curr = (curr + i) % m
  for prev in prefixes:
    temp_sum = (curr - prev) % m
    if temp_sum > max_sum:
      max_sum = temp_sum
  prefixes.append(curr)
print(max_sum)
```

```python
#W7A13
nums = list(map(int, input().split()))
freq = {}
for i in nums:
  if i not in freq:
    freq[i] = 1
  else:
    freq[i] += 1
max_len = 0
for nums in freq:
  curr_len = freq[nums]
  if nums + 1 in freq:
    curr_len += freq[nums + 1]
  if curr_len > max_len:
    max_len = curr_len
print(max_len)
```
