```python
#W9A1
def merge(nums1, nums2):
  i = 0
  j = 0
  res = []
  while i < len(nums1) and j < len(nums2):
    if nums1[i] < nums2[j]:
      res.append(nums1[i])
      i += 1
    else:
      res.append(nums2[j])
      j += 1
  while i < len(nums1):
    res.append(nums1[i])
    i += 1
  while j < len(nums2):
    res.append(nums2[j])
    j += 1
  return res
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*(merge(a, b)))
```

```python
#W9A3
n = input().strip()
lst_n = [int(x) for x in n]
length = len(lst_n)

if 0 in lst_n:
  idx_0 = n.rfind('0')
  lst_n[0], lst_n[idx_0] = lst_n[idx_0], lst_n[0]
else:
  swapped = False
  for i in range(length):
    r = lst_n[i+1:]
    if not r:
      break
    min_r = min(r)
    if min_r < lst_n[i]:
      for j in range (length-1, i, -1):
        if lst_n[j] == min_r:
          lst_n[i], lst_n[j] = lst_n[j], lst_n[i]
          swapped = True
          break
      if swapped:
        break
  if not swapped:
    lst_n[-1], lst_n[-2] = lst_n[-2], lst_n[-1]

res = ''.join(map(str, lst_n))
print(int(res))
```

```python
#W9A4
def solve(s1, s2):
  n1 = len(s1)
  n2 = len(s2)
  p1 = 0
  p2 = 0
  while p1 < n1:
    if s1[p1].upper() != s2[p2]:
      p1 += 1
    else:
      p1 += 1
      p2 += 1
    if p2 == n2:
      return True
  return False
n = int(input())
for i in range(n):
  a = input()
  b = input()
  print("YES" if solve(a, b) else "NO")
```

```python
#W9A5
def solve(lst1, lst2):
  n1 = len(lst1)
  n2 = len(lst2)
  if n1 == 0:
    return True
  if n1>n2:
    return False
  for i in range(n2-n1+1):
    if lst1 == lst2[i:i+n1]:
      return True
  return False
nums1 = list(map(int, input().split()))
nums2 = list(map(int, input().split()))
print(solve(nums1, nums2))
```

```python
#W9A6
def hoanvi(arr, n):
  i = n - 2
  while arr[i+1]<=arr[i]:
    i-=1
  j = n - 1
  while(arr[j] < arr[i]):
    j -= 1
  arr[i], arr[j] = arr[j], arr[i]
  ans = arr[0:i+1]+list(reversed(arr[i+1:]))
  return ans
n = int(input())
lst = list(map(int, input().split()))
print(*hoanvi(lst, n))
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
  count = 0
  n = len(flowerbed)
  for i in range(n):
    if flowerbed[i] == 0 and (i == 0 or flowerbed[i-1] == 0) and (i == n-1 or flowerbed[i+1] == 0):
      flowerbed[i] = 1
      count += 1
    if count >= k:
      return True
  return False

lst = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(lst, k))
```

```python
#W9A8
def is_path_crossing(moves):
  x, y = 0, 0
  di_qua = {(0, 0)}
  for move in moves:
    if move == 'R':
      x+=1
    if move == 'L':
      x-=1
    if move == 'U':
      y+=1
    if move == 'D':
      y-=1
    if (x, y) in di_qua:
      return True
    di_qua.add((x, y))
  return False
moves = input()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
  if nums[0] > nums[1]:
    nums[0], nums[1] = nums[1], nums[0]
  if nums[2] > nums[3]:
    nums[2], nums[3] = nums[3], nums[2]
  if nums[0] > nums[2]:
    nums[0], nums[2] = nums[2], nums[0]
    nums[1], nums[3] = nums[3], nums[1]
  if nums[1] > nums[4]:
    nums[1], nums[4] = nums[4], nums[1]
  if nums[1] > nums[2]:
    nums[1], nums[2] = nums[2], nums[1]
    nums[3], nums[4] = nums[4], nums[3]
  return nums[2] if nums[2] < nums[4] else nums[4]
nums = list(map(int, input().split()))
print(Median(nums))
```

```python
def findRadius(houses, heaters):
  for i in range(len(houses)):
    min_dis = -1
    for j in range (len(heaters)):
      dis = abs(houses[i] - heaters[j])
      if min_dis == -1 or dis < min_dis:
        min_dis = dis
    houses[i] = min_dis
  return max(houses)
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
