```python
#W8A3
def happynumbers(n):
  res = 0
  while (n > 0):
    res += (n%10)**2
    n//=10
  return res
def solve(x):
  seen = set()
  while x != 1 and x not in seen:
    seen.add(x)
    x = happynumbers(x)
  return x == 1
try:
  a = int(input())
  print("YES" if solve(a) else "NO")
except ValueError:
  print("Err")
```

```python
#W8A5
def solve(lst):
  for x in lst:
    try:
      x = int(x)
    except ValueError:
      print(f'{x} khong phai la so nguyen')
      continue
lst = list(input().split())
solve(lst)
```

```python
#8
def solve(s):
  if '.' not in s:
    raise ValueError("File khong hop le")
  lst_s = s.split(".")
  if len(lst_s) != 2:
    raise ValueError("File khong hop le")
  if lst_s[1] == 'txt' or lst_s[1] == 'zip':
    print("Doc file thanh cong")
  else:
    raise ValueError("File khong hop le")
try:
  s = input()
  solve(s)
except ValueError as e:
  print(e)
```

```python
#10
try:
  n = list(map(int, input().split()))
  if n[0] != len(n)-1:
    raise ValueError
  if len(set(n[1:])) != len(n) - 1:
    raise ValueError
  print(sum(n[1:]))
except ValueError:
  print("Mang khong hop le")
```
